# K-Nearest Neighbors (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import numpy as np
```

## KNN Example:

In this example, we will explore the application of K-Nearest Neighbors (KNN) for classifying neural firing rate data into two conditions: Rest and Task.

Our approach to analyzing the data is structured as follows:
1. **Training the KNN Classifier:** We will train a KNN model with $$k = 3$ using the provided dataset of 20 samples, which includes two features (firing rates from two brain regions) and corresponding condition labels.
2. **Predicting the Condition:** We will use the trained KNN model to predict the condition for a new sample with firing rates Firing rate 1 = 14, firing rate 2 = 9.
3. **Analyzing the Impact of k:** We will investigate how the choice of k affects the classification performance, highlighting the trade-offs between low and high values of k.

By working through this example, we will gain a deeper understanding of how KNN operates as an instance-based learning algorithm, how it uses distance metrics to classify new samples, and how the choice of k influences model sensitivity and robustness. This example will illustrate the role of KNN in neuroscience for classifying brain activity data and its implications for practical applications.

### Create synthetic dataset

```python
# Dataset
firing_rates = np.array([
    [12, 8], [15, 10], [13, 7], [14, 9], [16, 11],
    [11, 6], [13, 8], [15, 9], [10, 5], [17, 12],
    [14, 8], [9, 6], [13, 7], [16, 10], [12, 9],
    [15, 11], [10, 4], [14, 10], [11, 7], [13, 6]
])
labels = np.array([
    'Rest', 'Task', 'Rest', 'Task', 'Task',
    'Rest', 'Rest', 'Task', 'Rest', 'Task',
    'Task', 'Rest', 'Rest', 'Task', 'Rest',
    'Task', 'Rest', 'Task', 'Rest', 'Rest'
])
```

### Create the model

```python
# Create and train the KNN model
k = 3
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(firing_rates, labels)
```

### Predict the class of new sample

```python
# New sample
new_sample = np.array([[14, 9]])
# Predict the condition for the new sample
predicted_class = knn.predict(new_sample)
print(f"Predicted condition for the new sample: {predicted_class[0]}")
```

### Scatter plot

```python
# Scatter plot
plt.figure(figsize=(8, 6))

# Plot Rest points
rest_points = firing_rates[labels == 'Rest']
plt.scatter(rest_points[:, 0], rest_points[:, 1], color='blue', label='Rest', alpha=0.7)

# Plot Task points
task_points = firing_rates[labels == 'Task']
plt.scatter(task_points[:, 0], task_points[:, 1], color='orange', label='Task', alpha=0.7)

# Plot New Sample
plt.scatter(new_sample[0, 0], new_sample[0, 1], color='red', label='New Sample', edgecolor='black', s=100, marker='*')

# Add labels and legend
plt.title("Scatter Plot of Firing Rates with New Sample")
plt.xlabel("Firing Rate 1")
plt.ylabel("Firing Rate 2")
plt.legend(loc='upper left')
plt.grid(alpha=0.3)
plt.show()
```
