# Exercise 1 Solution: Recall of Probability

In this exercise, we explore fundamental probability concepts and their applications in neuroscience. Understanding probability is essential for analyzing and interpreting neural data, as it forms the foundation for statistical inference.

Specifically, we will:
1. **Cumulative Distribution Function (CDF) and EEG Signal Amplitude**: Compute and visualize the CDF of EEG signal amplitudes, which follow a normal distribution.
2. **Probability Mass Function (PMF) and Neuron Firing Counts**: Model the discrete probability distribution of neuronal spike counts using the Poisson distribution.
3. **Variance of Reaction Times**: Compute the variance of reaction times in a cognitive task, demonstrating how spread influences interpretation.

By the end of this exercise, we will have a comprehensive understanding of probability distributions and their relevance in neuroscience.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import poisson
from scipy.stats import uniform
from scipy.stats import expon
```

## Example 1: CDF and EEG Signal Amplitude Distribution

```python
"""
---- Compute and plot the CDF ----
"""
# Given parameters
mu = 50  # Mean EEG amplitude in µV
sigma = 10  # Standard deviation in µV

# Generate values for plotting
x = np.linspace(20, 80, 1000)
cdf_values = norm.cdf(x, loc=mu, scale=sigma)

# Plot the CDF
plt.plot(x, cdf_values, label="CDF of EEG Amplitude", color="blue")
plt.axvline(40, color="red", linestyle="--", label="Threshold at 40 µV")
plt.xlabel("EEG Amplitude (µV)")
plt.ylabel("Cumulative Probability")
plt.title("CDF of EEG Signal Amplitude")
plt.legend()
plt.grid()
plt.show()
```

```python
"""
The probability that X < 40
"""
p_40 = norm.cdf(40, loc=mu, scale=sigma)
print(f"Probability that EEG amplitude is less than 40 µV: {p_40:.4f} = {p_40*100: .2f}%")
```

## Example 2: PMF and Neuron Firing Counts

```python
# Poisson parameter
lambda_ = 5  # Average spikes per second

# Values for PMF
x = np.arange(0, 15)  # Possible spike counts
pmf_values = poisson.pmf(x, lambda_)

# Plot the PMF
plt.stem(x, pmf_values, linefmt='b-', markerfmt='bo', basefmt='r-')
plt.xlabel("Number of Spikes")
plt.ylabel("Probability Mass Function (PMF)")
plt.title("PMF of Neuron Firing")
plt.grid()
plt.show()
```

```python
# Probability of exactly 3 spikes
p_3 = poisson.pmf(3, lambda_)
print(f"Probability of exactly 3 spikes in 1 second: {p_3:.4f} = {p_3 * 100: .2f}%")
```

## Example 3: Expected Value in fMRI Signal Strength

```python
# Define the uniform distribution parameters
a, b = 20, 80  # Range of BOLD signal

# Expected value calculation
E_X = (a + b) / 2
print(f"Expected value of the BOLD signal: {E_X:.2f}")
```

```python
# Simulating 10,000 samples
samples = uniform.rvs(loc=a, scale=(b-a), size=10000)

# Compute and compare the sample mean
sample_mean = np.mean(samples)
print(f"Simulated mean from 10,000 samples: {sample_mean:.2f}")
```

## Example 4: Variance of Reaction Times

```python
# Given mean reaction time
mu = 250  # Mean RT in ms
lambda_ = 1 / mu  # Rate parameter

# Compute variance
Var_X = 1 / lambda_**2
print(f"Theoretical variance of RT: {Var_X:.2f} ms²")
```

```python
# Simulate 10,000 reaction times
simulated_rt = expon.rvs(scale=mu, size=10000)

# Compute and compare variance
simulated_variance = np.var(simulated_rt)
print(f"Simulated variance from 10,000 samples: {simulated_variance:.2f} ms²")
```

**Note**: The simulated data should be close to this value of the theoretical value.
